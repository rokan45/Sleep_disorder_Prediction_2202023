```python
import pandas as pd

dataset=pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
dataset.head()
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['Gender'] = lb.fit_transform(dataset['Gender'])
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['Occupation'] = lb.fit_transform(dataset['Occupation'])
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['BMI Category'] = lb.fit_transform(dataset['BMI Category'])
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['Blood Pressure'] = lb.fit_transform(dataset['Blood Pressure'])
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['Sleep Disorder'] = lb.fit_transform(dataset['Sleep Disorder'])
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset['Quality of Sleep'],dataset['Sleep Disorder'])
plt.title("Scatter Plot")
plt.xlabel('Quality of Sleep')
plt.ylabel('Sleep Disorder')
plt.show();
```

```python
x=dataset.drop(columns=['Sleep Disorder'])
y=dataset['Sleep Disorder']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(
    x,y,test_size=0.3,random_state=42
    )
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_squared_error as mse
predicted=lr.predict(xtest)
error=mse(ytest,predicted)**.5
print(error)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

models = [RandomForestRegressor(), DecisionTreeRegressor(), LinearRegression()]

errors = {}

for model in models:
    model.fit(xtrain,ytrain)
    name = type(model).__name__
    score = model.score(xtest,ytest)*100
    errors[name] = score
```
